## 1. Array creation

- Create a 1D array called ```xarray``` with values from 0 to 10 and in steps of 0.1. Check the shape of the array:

````python
xarray = np.arange(0, 10.1, 0.1)
xarray
````

````python
print('The shape of the array is', xarray.shape)
````

- Create an array of normally distributed numbers with mean $\mu=0$ and standard deviation $\sigma=0.5$. It should have 20 rows and as many columns as there are elements in ```xarray```. Call it ```normal_array```:

````python
normal_array = np.random.normal(loc = 0, scale = 0.5, size = (20, xarray.shape[0]))
````

````python
print('The shape of the array is', normal_array.shape)
````

- Check the type of ```normal_array```:

````python
normal_array.dtype
````

## 2. Array mathematics

- Using ```xarray``` as x-variable, create a new array ```yarray``` as y-variable using the function $y = 10* cos(x) * e^{-0.1x}$:

````python
yarray = 10*np.cos(xarray)*np.exp(-0.1*xarray)
yarray
````

- Create ```array_abs``` by taking the absolute value of ```yarray```:

````python
array_abs = np.abs(yarray)
array_abs
````

- Create a boolean array (logical array) where all positions $>0.3$ in ```array_abs``` are ```True``` and the others ```False```

````python
boolean_array = array_abs > 0.3
boolean_array
````

- Create a standard deviation projection along the second dimension (columns) of ```normal_array```. Check that the dimensions are the ones you expected. Also are the values around the value you expect?

````python
std_array = np.std(normal_array, axis = 1)
std_array
````

````python
print('I valori sono circa 2, come mi aspettavo, ovvero circa pari al parametro scale passato come input')
print('Numero di righe di normal_array: ', normal_array.shape[0], '< --- > Numero di elementi di std_array', std_array.shape[0])
````

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (8, 5))
ax.plot(xarray, yarray, label = 'xarray')
ax.plot(xarray, array_abs, label = 'np.abs(xarray)')
ax.plot(xarray, boolean_array, label = '1 if xarray > 0.3 else 0')
ax.legend()
plt.show()
````

- ```yarray``` represents a signal. Each line of ```normal_array``` represents a possible random noise for that signal. Using broadcasting, try to create an array of noisy versions of ```yarray``` using ```normal_array```.

````python
print(yarray.shape)
print(normal_array.shape)
````

````python
noisy_versions_of_yarray = normal_array + yarray
print(noisy_versions_of_yarray.shape)
````

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (20, 10))
for j in range(noisy_versions_of_yarray.shape[0]):
    ax.plot(xarray, noisy_versions_of_yarray[j,:], label = 'noisy version of yarray N' + str(j+1))
ax.plot(xarray, yarray, lw = 4, c = 'black', label = 'yarray')
ax.legend()
plt.show()
````
